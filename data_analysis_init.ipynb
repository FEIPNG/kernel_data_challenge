```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

# Load the datasets
file_x = "./data/Xtr0.csv"
file_y = "./data/Ytr0.csv"

df_x = pd.read_csv(file_x)
df_y = pd.read_csv(file_y)

# Merge datasets on 'Id'
df = df_x.merge(df_y, on='Id')

# Dataset overview
print("Dataset Info:\n")
print(df.info())

# Check for missing values
print("\nMissing Values:\n", df.isnull().sum())

# Sequence length distribution
df['seq_length'] = df['seq'].apply(len)
print("\n")
print(df["seq_length"].value_counts())
print("\n")

# Nucleotide composition
nucleotides = "ACGT"
all_sequences = "".join(df['seq'])
nucleotide_counts = Counter(all_sequences)
nucleotide_freq = {nt: nucleotide_counts[nt] / len(all_sequences) for nt in nucleotides}

# print("\nNucleotide Frequencies:")
# for nt, freq in nucleotide_freq.items():
#     print(f"{nt}: {freq:.4f}")

# Plot nucleotide frequencies
plt.figure(figsize=(6, 4))
sns.barplot(x=list(nucleotide_freq.keys()), y=list(nucleotide_freq.values()), palette='viridis')
plt.title("Nucleotide Frequencies")
plt.xlabel("Nucleotide")
plt.ylabel("Frequency")
plt.show()

# Class balance
plt.figure(figsize=(6, 4))
sns.countplot(x='Bound', data=df, palette='coolwarm')
plt.title("Class Distribution (Bound vs. Unbound)")
plt.xlabel("Bound")
plt.ylabel("Count")
plt.show()

print(df["Bound"].value_counts())
```
